## Zad 1.
Zaimplementuj procedurę Sinkhorna-Knoppa dla problemu transportowego optymalnego. Dany jest:
- wektor źródeł: $a = (0.2, 0.3, 0.5)^T$,
- wektor ujść: $b = (0.4, 0.1, 0.5)^T$.

Macierz kosztów:
$$
C = \begin{bmatrix}
1 & 2 & 3 \\
2 & 1 & 2 \\
3 & 2 & 1
\end{bmatrix}
$$
Znajdź przybliżony plan transportowy używając metody *Sinkhorna-Knoppa* z parametrem regularyzacji $\varepsilon = 0.1$.
